````python
import torch

from transformers import AutoModelForCausalLM, AutoTokenizer
from datasets import load_dataset

from IPython.display import Markdown, display

from dotenv import load_dotenv

load_dotenv('../.env')
````

````python
device = 'cuda' if torch.cuda.is_available() else 'mps' if torch.mps.is_available() else 'cpu'
````

````python
def generate(model: AutoModelForCausalLM, tokenizer, prompt, **gen_kwargs) -> str:
    inputs = tokenizer(prompt, return_tensors='pt')
    inputs = {k:v.to(device) for k,v in inputs.items()}
    outputs = model.generate(**inputs, **gen_kwargs).squeeze(0).cpu()
    response = tokenizer.decode(outputs)
    return response
````

````python
dataset = load_dataset('openai/gsm8k', 'main', split='train')

sample = dataset[0]
````

````python
models = {
    'pythia_70m': 'EleutherAI/pythia-70m',
    'smollm_135m_it': 'HuggingFaceTB/SmolLM-135M-Instruct',
    'smollm2_135m_it': 'HuggingFaceTB/SmolLM2-135M-Instruct',
    'pythia_160m': 'EleutherAI/pythia-160m',
    'gemma3_270m_it': 'google/gemma-3-270m-it',
    'smollm_360m_it': 'HuggingFaceTB/SmolLM-360M-Instruct',
    'smollm2_360m_it': 'HuggingFaceTB/SmolLM2-360M-Instruct',
    'qwen3_0.6b': 'Qwen/Qwen3-0.6B'
}
````

````python
question = sample['question']
answer = sample['answer']

display(Markdown("## Question"))
display(Markdown(question))
display(Markdown("## Answer"))
display(Markdown(answer))
display(Markdown('---'))

for model_name, model_path in models.items():
    model = AutoModelForCausalLM.from_pretrained(model_path, device_map=device)
    tokenizer = AutoTokenizer.from_pretrained(model_path)
    tokenizer.pad_token = tokenizer.eos_token

    try:
        prompt = tokenizer.apply_chat_template([{'role': 'system', 'content': 'You are a helpful assistant that answers questions correctly'},{'role': 'user', 'content': question}], tokenize=False, enable_thinking=False, add_generation_prompt=True)
    except:
        print(f"WARNING: NO CHAT TEMPLATE FOR {model_name}")
        prompt = f"Question: {question}\nAnswer: "

    display(Markdown(f'## {model_name}'))
    display(Markdown('#### Prompt'))
    display(Markdown(f"""```\n{prompt}\n```"""))

    response = generate(model, tokenizer, prompt, max_new_tokens=256, temperature=1.0, do_sample=True)
    display(Markdown('#### Response'))
    display(Markdown(f"""```\n{response}\n```"""))
    display(Markdown('---'))
````
