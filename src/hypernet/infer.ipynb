```python
from dataset import load_dataset

from hypernetwork import TaskWeaver
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
hn = TaskWeaver.from_pretrained('./taskweaver_output').to('cuda')
```

```python
model = AutoModelForCausalLM.from_pretrained(hn.model_name).to('cuda')
tokenizer = AutoTokenizer.from_pretrained(hn.model_name)
```

```python
dataset = load_dataset('openai/gsm8k', 'main')
```

```python
prompt = dataset['train'][0]['question']
print(prompt)

inputs = tokenizer(prompt, return_tensors='pt')
inputs = {k:v.to('cuda') for k,v in inputs.items()}
```

```python
model_output = model.generate(**inputs, max_new_tokens=512).squeeze()
model_response = tokenizer.decode(model_output)
print(model_response)
```

```python
hn_output = hn.generate(**inputs, max_new_tokens=512).squeeze()
hn_response = tokenizer.decode(hn_output)
print(hn_response)
```
